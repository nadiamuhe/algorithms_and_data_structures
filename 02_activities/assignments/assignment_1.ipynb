## Question Two: Valid Bracket Sequence
**Description**  
Given a string containing only the characters `'('`, `')'`, `'{'`, `'}'`, `'['`, and `']'`, determine if the input string is a **valid bracket sequence**.  
A string is valid if:
- Open brackets are closed by the same type of brackets, and
- Open brackets are closed in the correct order.

**Examples**
```python
Input: s = "([]{})"
Output: True
```
```python
Input: s = "([)]"
Output: False
```
```python
Input: s = "()[]{}"
Output: True
```
```python
Input: s = "[{]}"
Output: False
```

**Question 2 Starter Code**

````python
def is_valid_brackets(s: str) -> bool:
    opening = ['(', '[', '{']
    closing = [')', ']', '}']
    stack = []
    for char in s:
        if char in opening:
            stack.append(char)
        elif char in closing:
            if len(stack) > 0 and opening.index(stack[-1]) == closing.index(char):
                stack.pop()
            else:
                return False

    return not stack
````

## Question Three: Move All Zeros to End
**Description**  
Given a list of integers, move all zeros to the end while maintaining the relative order of the non-zero elements. 

**Examples**
```python
Input: nums = [0, 1, 0, 3, 12]
Output: [1, 3, 12, 0, 0]
```
```python
Input: nums = [4, 0, 5, 0, 0, 6]
Output: [4, 5, 6, 0, 0, 0]
```


````python
from typing import List

def move_zeros_to_end(nums: List[int]) -> List[int]:
    insert_pos = 0
    for num in nums:
        if num != 0:
            nums[insert_pos] = num
            insert_pos += 1
    for i in range(insert_pos, len(nums)):
        nums[i] = 0
    return nums
    
````

````python
from typing import List

def move_zeros_to_end(nums: List[int]) -> List[int]:
    insert_pos = 0
    for num in nums:
        if num != 0:
            nums[insert_pos] = num
            insert_pos += 1
    for i in range(insert_pos, len(nums)):
        nums[i] = 0
    return nums
````


-   Explain why your solution works


My solution works because it keeps the order of the non-zero elements and accounts for edge cases. The solution requires a pointer, insert_pos, that keeps track of the next position where a non-zero element should go. It loops over the list once and every time it reaches a non-zero element, it places that element at the insert_pos position and moves the pointer forward by one. Because the list is read from left to right and only non-zero elements are written, their relative order is maintained. After this loop, all the non-zero elements are at the front and insert_pos marks where they end. Then a second loop fills every remaining position from insert_pos to the end with zeros. The code also accounts for edge cases where the list is all zeros, in which case the pointer never moves and the whole list is filled with zeros, or where the list has no zeros, in which case the pointer reaches the end and the second loop does nothing.


-   Explain the problem’s time and space complexity


Time: O(n). The solution loops over the list once to move the non-zero elements, then loops again to fill in the zeros. Each operation inside the loops costs O(1) per iteration. Therefore the total time would be O(n).
Space: O(1). The list is modified in place and the solution only uses a constant number of extra variables regardless of the input size. No additional structure grows so the memory used is O(1).


-   Explain the thinking to an alternative solution (no coding required, but a classmate reading this should be able to code it up based off your text)


The alternative solution would be to build a new list instead of modifying the list in place. We create an empty list and a counter set to zero. Then we loop over the original list and check each element. If the element is non-zero, we append it to the new list. If the element is a zero, we increment the counter instead of appending. After the loop, the new list holds all the non-zero elements in their original order. Then we append zeros to the end of the new list by adding the counter's worth of zeros, for example by concatenating [0] * counter. We return the new list. This solution uses O(n) space because it creates a separate list the same size as the original instead of reusing it.
